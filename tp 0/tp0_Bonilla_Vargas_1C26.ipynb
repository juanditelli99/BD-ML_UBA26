```python
# Resolver acá

listado_años = [2026, 2027, 2028]
for numero in listado_años:
    if numero % 400 == 0: 
        print(f"{numero} es un año bisiesto por 1era condición")
    elif numero % 4 == 0 and numero % 100 != 0:
        print(f"{numero} es un año bisiesto por 2da condición")
    else:
        print(f"{numero} no es un año bisiesto")
```

```python
# Resolver acá

# Valor de referencia: proporcionalidad
argentina = 46044703 / 257

# Creamos variables para una selección de jurisdicciones
cordoba = 3978984 / 18
santa_fe = 3556522 / 19
mendoza = 2014533 / 10
buenos_aires = 17569053 / 70
entre_rios =  1426426 / 9
santa_cruz = 333473 / 5
formosa = 606041 / 5

# Definimos el diccionario para iterar
dict_provincias = {
    "Córdoba": cordoba,
    "Santa Fe": santa_fe,
    "Mendoza": mendoza,
    "Buenos Aires": buenos_aires,
    "Entre Ríos": entre_rios,
    "Santa Cruz": santa_cruz,
    "Formosa": formosa
}

mayor_sobrerep = None      # en esta va guardo el nombre de la provincia
valor_sobrerep = argentina # valor de comparación, cualquiera sobrerep tiene que ser menor a argentina

for provincia, valor in dict_provincias.items():
    if valor < argentina:
        print(f"{provincia} está sobrerepresentada: {valor:.0f} hab/diputado") #el .0f es para mostrar 0 decimales del float (núm. decimal)
        # ¿es la más sobrerepresentada hasta ahora? esto se pregunta y guarda con cada provincia que va tomando del diccionario
        if valor < valor_sobrerep:
            valor_sobrerep = valor
            mayor_sobrerep = provincia
    elif valor > argentina:
        print(f"{provincia} está subrepresentada: {valor:.0f} hab/diputado")
    else:
        print(f"{provincia} tiene representación proporcional")

print(f"\nProvincia más sobrerepresentada: {mayor_sobrerep} con {valor_sobrerep:.0f} hab/diputado")   #el \n es como un \\ en latex, para dar un salto de línea.
```

```python
# Estos son los valores de la provincia
representantes_pba = 70
poblacion2010_pba = 17594428


# Y el valor de referencia nacional
argentina = 46044703 / 257
```

```python
# Resolver acá

# Estos son los valores de la provincia
representantes_pba = 70
poblacion2010_pba = 17594428


# Y el valor de referencia nacional
argentina = 46044703 / 257

pba = poblacion2010_pba / representantes_pba
pba_ideal = 17594428 / argentina
print(f"PBA tiene {pba:.0f} habitantes/diputado") 
print(f"La proporcionalidad exacta exigiría {argentina:.0f} habitantes/diputado")
print(f"\nEntonces PBA ideal tendría {pba_ideal:.0f} diputados y no los {representantes_pba} actuales")

while representantes_pba < pba_ideal:
    print(f"Se agregó un representante a los {representantes_pba} actuales, el total ahora es {pba_ideal:.0f} diputados")
    representantes_pba += 1
    
```
